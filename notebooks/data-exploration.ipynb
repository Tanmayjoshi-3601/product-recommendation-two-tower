## 4.Split into Train/Test/Val (By User)

```python
rng = np.random.default_rng(42)

users = data["user_id_enc"].unique()
rng.shuffle(users)

n = len(users)
train_users = set(users[:int(0.8*n)])
val_users = set(users[int(0.8*n):int(0.9*n)])
test_users = set(users[int(0.9*n):])

train = data[data['user_id_enc'].isin(train_users)].copy()
val = data[data['user_id_enc'].isin(val_users)].copy()
test = data[data['user_id_enc'].isin(test_users)].copy()

print(len(train),len(val), len(test))
```

## 5.Handling Data Imbalance - Downsampling positives per user (train only)

```python
def balance_dataset_global(
    df: pd.DataFrame,
    label_col: str = "label",
    method: str = "downsample",  # or "hybrid"
    target_ratio: float = 1.0,
    random_state: int = 42
) -> pd.DataFrame:
    """
    Global balancing - simpler and often more effective for two-tower models
    """
    pos = df[df[label_col] == 1]
    neg = df[df[label_col] == 0]
    
    n_pos = len(pos)
    n_neg = len(neg)
    
    print(f"Original: {n_pos} positives, {n_neg} negatives")
    
    if method == "downsample":
        # Downsample majority class to match minority
        if n_pos > n_neg:
            n_target_pos = int(n_neg * target_ratio)
            pos_sampled = pos.sample(n=n_target_pos, random_state=random_state)
            result = pd.concat([pos_sampled, neg], axis=0)
        else:
            n_target_neg = int(n_pos / target_ratio)
            neg_sampled = neg.sample(n=n_target_neg, random_state=random_state)
            result = pd.concat([pos, neg_sampled], axis=0)
    
    elif method == "hybrid":
        # Downsample majority and upsample minority slightly
        target_total = int((n_pos + n_neg) * 0.6)  # Keep 60% of data
        n_target_each = target_total // 2
        
        # Sample with replacement if needed
        pos_sampled = pos.sample(n=n_target_each, 
                                replace=(n_target_each > n_pos),
                                random_state=random_state)
        neg_sampled = neg.sample(n=n_target_each,
                                replace=(n_target_each > n_neg), 
                                random_state=random_state)
        result = pd.concat([pos_sampled, neg_sampled], axis=0)
    
    # Shuffle
    result = result.sample(frac=1.0, random_state=random_state).reset_index(drop=True)
    
    print(f"Balanced: {len(result[result[label_col]==1])} positives, {len(result[result[label_col]==0])} negatives")
    print(f"Final distribution:\n{result[label_col].value_counts(normalize=True)}")
    
    return result

# Use global balancing
train_bal = balance_dataset_global(
    train,
    method="downsample",  # Simple and effective
    target_ratio=1.0,      # 1:1 ratio
    random_state=42
)
```

```python
# Check what subclasses exist in your positives
def analyze_subclass_distribution(df, label_col='label'):
    positives = df[df[label_col] == 1]
    
    print("=== Subclass Distribution in Positives ===")
    
    # By rating value
    print("\nRating distribution (4 vs 5 stars):")
    print(positives['rating'].value_counts(normalize=True))
    
    # By product category (if you have it)
    if 'main_category' in positives.columns:
        print("\nCategory distribution:")
        print(positives['main_category'].value_counts(normalize=True))
    
    # By price range
    if 'price' in positives.columns:
        positives['price_range'] = pd.cut(positives['price'], 
                                         bins=[0, 10, 25, 50, 100, float('inf')],
                                         labels=['<$10', '$10-25', '$25-50', '$50-100', '>$100'])
        print("\nPrice range distribution:")
        print(positives['price_range'].value_counts(normalize=True))
    
    # By product popularity (review count as proxy)
    positives['popularity'] = pd.qcut(positives['rating_number'], 
                                     q=[0, .25, .75, 1.0],
                                     labels=['Niche', 'Medium', 'Popular'])
    print("\nProduct popularity:")
    print(positives['popularity'].value_counts(normalize=True))
    
    return positives

# Analyze before and after sampling
original_dist = analyze_subclass_distribution(train)
balanced_dist = analyze_subclass_distribution(train_bal)
```

```python
def compare_sampling_methods(train_df, n_iterations=10):
    """
    Compare random vs stratified sampling stability
    """
    random_variances = []
    stratified_variances = []
    
    for i in range(n_iterations):
        # Random sampling
        random_sample = train_df.sample(n=34600, random_state=i)
        random_dist = random_sample['rating'].value_counts(normalize=True)
        random_variances.append(random_dist.values)
        
        # Stratified sampling
        from sklearn.model_selection import train_test_split
        strat_sample, _ = train_test_split(
            train_df, 
            train_size=34600,
            stratify=train_df['rating'],
            random_state=i
        )
        strat_dist = strat_sample['rating'].value_counts(normalize=True)
        stratified_variances.append(strat_dist.values)
    
    print("Variance across multiple runs:")
    print(f"Random sampling std: {np.std(random_variances):.6f}")
    print(f"Stratified sampling std: {np.std(stratified_variances):.6f}")
```

```python
compare_sampling_methods(train)
```
